# Quick Sort where main element is the first

```
def partition(array, left, right):

  x = array[left]
  j = left
  for i in range(left + 1, right):
    if array[i] <= x:
      array[i], array[j + 1] = array[j + 1], array[i]
      j += 1
  
  array[left], array[j] = array[j], array[left]

  return j

def quickSort(array, left, right):

  if left >= right:
    return

  j = partition(array, left, right)
  quickSort(array, left, j)
  quickSort(array, j + 1, right)
```

```
import random
import time
n = 10**4
lst = [random.randint(0, 100) for i in range(n)]
print(lst)
start = time.perf_counter()
quickSort(lst, 0, len(lst))
end = time.perf_counter()
print(end - start)
```

# Quick Sort where main element is a random

```
import random


def partition(array, left, right):

  random_index = random.randint(left, right - 1)
  x = array[random_index]
  array[random_index], array[left] = array[left], array[random_index]
  j = left

  for i in range(left + 1, right):
    if array[i] <= x:
      array[i], array[j + 1] = array[j + 1], array[i]
      j += 1
  
  array[left], array[j] = array[j], array[left]

  return j

def quickSort(array, left, right):

  if left >= right:
    return

  j = partition(array, left, right)
  quickSort(array, left, j)
  quickSort(array, j + 1, right)
```

```
import time

lst = [random.randint(0, 100) for i in range(10)]
print(lst)
start = time.perf_counter()
quickSort(lst, 0, len(lst))
end = time.perf_counter()
print(end - start)
```

# Quick Sort with extra zone (when we have a lot of the same value)

```
def partition(array, left, right):
  
  random_index = random.randint(left, right - 1)
  x = array[random_index]
  array[random_index], array[left] = array[left], array[random_index]
  j = left

  for i in range(left + 1, right):
    if array[i] <= x:
      array[i], array[j + 1] = array[j + 1], array[i]
      j += 1
  
  diff_index = left

  for k in range(left, j + 1):
    if array[k] < x:
      array[k], array[diff_index + 1] = array[diff_index + 1], array[k]
      diff_index += 1
  
  array[left], array[diff_index] = array[diff_index], array[left]

  return diff_index  


def quickSort3(array, left, right):

  if left >= right:
    return

  j = partition(array, left, right)
  quickSort3(array, left, j)
  quickSort3(array, j + 1, right)
```

```
import time

lst = [random.randint(0, 100) for i in range(10)]
print(lst)

start = time.perf_counter()

quickSort3(lst, 0, len(lst))

end = time.perf_counter()
print(end - start)

print(lst)
```

```
a = [1, 2, 3, 4, 4, 4, 5, 6, 4, 6, 2, 1, 4]
b = [1, 2, 3, 4, 4, 4, 5, 6, 4, 6, 2, 1, 4]

print('First array before sort: ', a)
print('Second array before sort: ', b)

a_start = time.perf_counter()

quickSort(a, 0, len(a))

a_end = time.perf_counter()

b_start = time.perf_counter()

quickSort3(b, 0, len(b))

b_end = time.perf_counter()

print('First array after sort: ', a)
print('Second array after sort: ', b)
print('Time execution for first array sorting = ', a_end - a_start)
print('Time execution for Second array sorting = ', b_end - b_start)

```
